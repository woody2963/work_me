```python
mnames=['movie_id','title','genres']
movies=pd.read_table(r'C:\Users\24784\Desktop\Github\pydata-book\ch02\movielens\movies.dat',sep='::',names=mnames)
movies[:5]
```

```python
movies['genres'].value_counts()
```

```python
rnames=['user_id','movie_id','rating','timestamp']
ratings=pd.read_table(r'C:\Users\24784\Desktop\Github\pydata-book\ch02\movielens\ratings.dat',sep='::',names=rnames)
ratings[:5]
```

```python
ratings['rating'].value_counts()
```

```python
data=pd.merge(pd.merge(users,ratings),movies)
data
```

```python
mean_ratings=data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
mean_ratings[:10]
```

```python
ratings_by_title=data['title'].value_counts()
ratings_by_title[:10]
```

```python
rating_by_title2=data.groupby('title').size()
rating_by_title2[:10]
```

```python
active_titles=rating_by_title2.index[rating_by_title2>=250]
active_titles
```

```python
mean_ratings_active=mean_ratings.loc[active_titles]
mean_ratings_active[:10]
top_female_ratings=mean_ratings_active.sort_values(by='F',ascending=False)
top_female_ratings[:10]
```

```python
import matplotlib.pyplot as plt
top_female_ratings[:10].plot(kind='barh',title='Most popular movies Top 10')
plt.show()
```

```python
mean_ratings['diff']=mean_ratings['M']-mean_ratings['F']
mean_ratings
```

```python
mean_ratings.sort_values(by='diff')
```

```python
ratings_std=data.groupby('title')['rating'].std()
ratings_std=ratings_std.loc[active_titles]
ratings_std.sort_values(ascending=False)[:10]
```

```python
ratings_std2=data.pivot_table('rating',index='title',aggfunc='std')
ratings_std2=ratings_std2.loc[active_titles]
ratings_std2.sort_values(by='rating',ascending=False)[:10]
```
